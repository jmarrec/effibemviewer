```python
from pathlib import Path
import json
from IPython.display import HTML

# For geometry diagnostics to be included in GLTF, I have modified OpenStudio SDK source code
import sys
sys.path.insert(0, "/Users/julien/Software/Others/OS-build-release/Products/python")
import openstudio
sys.path.pop(0)


from effibemviewer import model_to_gltf_script, display_model, create_example_model

openstudio.openStudioLongVersion()
```

```python
# openstudio.Logger.instance().standardOutLogger().setLogLevel(openstudio.Debug)

model = create_example_model(include_geometry_diagnostics=True)
```

```python
ft = openstudio.model.ThreeJSForwardTranslator()
ft.setIncludeGeometryDiagnostics(False)
scene = ft.modelToThreeJS(model, True)
json = scene.toJSON(True)
Path("model.threeJS.json").write_text(json)

ft.setIncludeGeometryDiagnostics(True)
scene = ft.modelToThreeJS(model, True)
json = scene.toJSON(True)
Path("model_diags.threeJS.json").write_text(json)
```

```python
ft = openstudio.gltf.GltfForwardTranslator()
ft.setIncludeGeometryDiagnostics(False)
ft.modelToGLTF(model, "model.gltf")

ft = openstudio.gltf.GltfForwardTranslator()
ft.setIncludeGeometryDiagnostics(True)
ft.modelToGLTF(model, "model_diags.gltf")
```
