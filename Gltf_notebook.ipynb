```python
from pathlib import Path
import json
from IPython.display import HTML

# For geometry diagnostics to be included in GLTF, I have modified OpenStudio SDK source code
import sys
sys.path.insert(0, str(Path("~/Software/Others/OS-build-release/Products/python").expanduser()))
import openstudio
sys.path.pop(0)


from effibemviewer import display_model, create_example_model

openstudio.openStudioLongVersion()
```

```python
# openstudio.Logger.instance().standardOutLogger().setLogLevel(openstudio.Debug)

model = create_example_model(include_geometry_diagnostics=True)
```

```python
debug_dir = Path('LocalOnly')
debug_dir.mkdir(exist_ok=True)
ft = openstudio.gltf.GltfForwardTranslator()
ft.setIncludeGeometryDiagnostics(False)
ft.modelToGLTF(model, debug_dir / "model.gltf")

ft = openstudio.gltf.GltfForwardTranslator()
ft.setIncludeGeometryDiagnostics(True)
ft.modelToGLTF(model, debug_dir / "model_diags.gltf")
```
